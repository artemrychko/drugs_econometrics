```python
# fuction to parse of products' data
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys 
import csv
import re

def parsing(url):
    qwes = []
    for i in url:
        index_prod = str(i).replace('product', '').replace('/', '')
        try:
            asd = s.get('http://hydraruzxpnew4af.onion'+i)
            #asd = chrome.page_source
            soup = BeautifulSoup(asd.text, 'lxml')
            try:
                rating_shop = soup.find('div', 'rating_star')['title'].replace(' ', '').replace('\t', '').replace(' ', '')
            except:
                rating_shop = 'NA'
            try:
                rating_prod = soup.find('div', class_='product_rate').text.strip().replace('', '').replace('\n', '').replace('\t', '')
            except:
                rating_prod = 'NA'
            try:
                des_prod = soup.find('div', class_='text description').text.strip().replace('\n', '')
                des_n = len(str(des_prod))
            except:
                des_n = "NA"
                des_prod = "NA"
            try:
                city_prod = soup.find('ul', class_='meta text-muted i_location').find_all('li')
                city_prods = []
                for city in city_prod:
                    city_prods.append(city.text.strip())            
            except:
                city_prod = 'NA'
            try:    
                title_shop = soup.find('h1', class_='title over').text.strip()
            except:
                title_shop = 'NA'
            try:
                title_prod = soup.find('ol', class_='breadcrumb').text.strip().replace('', '').replace(title_shop,'').replace('\n', '')
            except:
                titel_prod = "NA"
            try:    
                price_prod = re.sub('.*','' ,str(soup.find('div', class_='product_price__big').text.strip())).replace('', '').replace(' ','')  
            except:
                price_prod = 'NA'
            try:    
                weight_prod = re.sub('.*/','' ,str(soup.find('div', class_='product_price__big').text.strip())).replace(' ', '') 
            except:
                weigth_prod = 'NA'
            try:
                dest_prod = soup.find_all('div', class_='nojs-tabs-body')
                dests_prod = []
                for dest in dest_prod:
                    opt = []
                    sdf = dest.find_all('li')
                    for sd in sdf:
                        city_type = []
                        weight = []
                        price_d = []
                        city_type.append(sd.find('div', class_='title').text.strip())
                        weight.append(sd.find('div', class_='av_nal').text.strip().replace('\xa0', ' '))
                        price_d.append(re.sub('\.*','', str(sd.find('div', class_='av_price').find('b'))).replace('<b>', '').replace(' ', ''))
                        opt.append([city_type, weight, price_d])
                    dests_prod.append(opt)
                    dests_n = len(opt)

            except:
                dests_prod = 'NA'
            try:
                comment_prod = soup.find_all('div', class_='comment')
                comments_prod = []
                for comm in comment_prod:
                        com_user = []
                        com_time = []
                        com_city = []
                        com_rating = []
                        com_comment = []
                        com_comment.append(comm.find('div', class_='text').text.strip())
                        try:
                            com_rating.append(comm.find('div', class_='rating_star')['title'].text.strip())
                        except:
                            com_rating = 'NA'
                        com_city.append(comm.find('i', class_='text-info pull-right').string)
                        asd1 = str(comm.find('span', class_='date text-muted').string).replace('', '25-11-2017').replace(' ', ';').replace('', '24-11-2017')
                        com_time.append(asd1)
                        com_user.append(comm.find('a', class_='name').string)
                        comments_prod.append([com_time, com_user, com_city, com_rating, com_comment])
            except:
                commnets_prod = 'NA'
            try:
                lis = soup.find('div', 'product_img_list').find_all('li')
                nn = []
                for li in lis:
                    nn.append(li.string)
                photos_prod = len(nn)
            except:
                photos_prod = 'NA'
        except:
            pass
        h1 = rating_shop    
        h2 = rating_prod
        h3 = des_prod
        h4 = city_prods
        h5 = title_shop
        h6 = title_prod
        h7 = price_prod
        h8 = dests_prod
        h9 = comments_prod
        h10 = weight_prod
        h11 = index_prod
        h12 = photos_prod
        h13 = dests_n
        h14 = des_n
        data = {"n_description":h14, 'n_dests':h13,'id_p':h11, 'rating_p':h2, 'description':h3, 'city_p':h4, 'title_p':h6, 'price_p':h7, "weight":h10,'destinations':h8, 'rating_s':h1, 'title_s':h5, 'n_photots':h12, 'comments':h9}
        qwes.append(data)
    return qwes
```
